# Snapshot
Snapshot is a present stock, future, option info. It contain open, high, low, close, change price, average price, volume, total volume, buy price, buy volume, sell price, sell volume and yestoday volume.<br>
snapshot函數能讓我們收取程式執行當下的商品資料，像是開高低收、總量、買賣價量等資料。<br>
也因為是抓取當下的資訊，所以Snapshots呼叫時無法給定特定的時間點，只須傳入Contract的List，而回傳的資料為shioaji.data.Snapshot物件的List。而Snapshots除了可以拿來抓取個股當下的交易資訊，也可以抓取期貨、選擇權或指數當下的交易資訊，只要把對應的Contract加入List即可。

## [First, you need login.](shioaji-quick_start.ipynb)

```python
import shioaji as sj
import getpass as gp

api = sj.Shioaji()
user_id, user_pass = gp.getpass("USER ID: "), gp.getpass("USER PASS: ")
accounts = api.login(user_id, user_pass)
```

## Get Snapshot

```python
contracts = [api.Contracts.Stocks['2330'],api.Contracts.Stocks['2317']]
snapshots = api.snapshots(contracts)
snapshots
```

## Snapshot to DataFrame

```python
import pandas as pd
contracts = [api.Contracts.Stocks['2330'],api.Contracts.Stocks['2317']]
snapshots = api.snapshots(contracts)
df = pd.DataFrame(snapshots)
df.ts = pd.to_datetime(df.ts)
df
```

> <details><summary>Attributes:</summary>
>
> ts (int): TimeStamp.<br>
> code (str): Contract id.<br>
> exchange (Exchange): Attributes of industry.<br>
> open (float): open<br>
> high (float): high<br>
> low (float): low<br>
> close (float): close<br>
> tick_type (TickType): Close is buy or sell price.<br>
>     {None, Buy, Sell}<br>
> change_price (float): change price.<br>
> change_rate (float): change rate.<br>
> change_type (ChangeType):<br>
>     {LimitUp, Up, Unchanged, Dowm, LimitDown}<br>
> avgerage_price (float): avgerage of price.<br>
> volume (int): volume.<br>
> total_volume (int): total volume.<br>
> amount (int): Deal amount.<br>
> total_amount (int): Total deal amount.<br>
> yestoday_volume (float): Volume of yestoday.<br>
> buy_price (float): Price of buy.<br>
> buy_volume (float): Volume of sell.<br>
> sell_price (float): Price of sell.<br>
> sell_volume (int): Volume of sell.<br>
> volume_ratio (float): total_volume/yestoday_volume.<br>
</details>
