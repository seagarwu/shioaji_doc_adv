# Profit Loss
Query account of profit loss 查詢已實現損益

## [First, you need login.](shioaji-quick_start.ipynb)

```python
import shioaji as sj
import getpass as gp

api = sj.Shioaji()
user_id, user_pass = gp.getpass("USER ID: "), gp.getpass("USER PASS: ")
accounts = api.login(user_id, user_pass)
```

```python
api.list_profit_loss?
```

```python
api.list_profit_loss(api.stock_account,'2021-11-01','2022-04-30')
```

```python
# 使用api.list_profit_loss_detail查詢已實現損益，但不知道為什麼都沒有顯示?!
api.set_default_account(accounts[0])
print(api.futopt_account)
profitloss = api.list_profit_loss(api.futopt_account,'2021-12-01','2022-04-30')
print(profitloss)
api.list_profit_loss_detail(api.futopt_account,0)
```

```python
# 反而得用api.get_account_settle_profitloss才看得到
import pandas as pd
api.set_default_account(accounts[2])
settle_profitloss = api.get_account_settle_profitloss(start_date='20211201',end_date='20220430')
df_profitloss = pd.DataFrame(settle_profitloss.data())
df_profitloss
```

> <details><summary>Attributes:</summary>
>
> id (int): use to find detail.<br>
> code (str): contract id.<br>
> seqno (str): seqno no.<br>
> dseq (str): seqno no.<br>
> quantity (int): quantity.<br>
> price (float): price.<br>
> pnl (float): profit and loss.<br>
> pr_ratio (float): profit rate.<br>
> cond (StockOrderCond): {Cash, Netting, MarginTrading, ShortSelling}<br>
> date (str): trade date.<br>
</details>
