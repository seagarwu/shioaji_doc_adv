# Short Stock Source

## [First, you need login.](shioaji-quick_start.ipynb)

```python
import shioaji as sj
import getpass as gp

api = sj.Shioaji()
user_id, user_pass = gp.getpass("USER ID: "), gp.getpass("USER PASS: ")
accounts = api.login(user_id, user_pass)
```

## Get Short Stock Source
**可用券源張數 (可融券的張數)**<br>
提供查詢每一檔股票的券源數量。值得注意的是，查詢到的結果，是自己這個帳戶歸屬的證券分點(分公司)的可用張數喔，並不是整個證券公司，更不是整個市場的數量。如果對融券、借券規則有不清楚的，一樣請先連絡您的營業員。<br>
這個數量指的就是現沖先賣(先賣後買)，券商目前所提供的額度上限。
而這個在iLeader中也有提到，若或有券張數不夠，在現沖先賣時就會跳出錯誤訊息。
https://www.sinotrade.com.tw/ec/140630/3.pdf<br>
確認券源，是我們執行放空操作或者價差配對交易的時候的重要先決，如果未來程式策略要自動作空，就可以先抓取當下券源數字，作為空單是否能執行的前置判斷。

```python
contracts = [api.Contracts.Stocks['2330'],api.Contracts.Stocks['2317']]
short_stock_sources = api.short_stock_sources(contracts)
short_stock_sources
```

## Short Stock Source to DataFrame

```python
import pandas as pd
contracts = [api.Contracts.Stocks['2330'],api.Contracts.Stocks['2317']]
short_stock_sources = api.short_stock_sources(contracts)
df = pd.DataFrame(short_stock_sources)
df.ts = pd.to_datetime(df.ts)
df
```

> Attributes:<br>
>
> code (str): Contract id.<br>
> short_stock_source (float): short_stock_source<br>
> ts (int): TimeStamp.<br>
