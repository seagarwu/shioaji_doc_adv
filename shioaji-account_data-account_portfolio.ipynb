## Get Settle ProfitLoss
取得平倉損益

````python
api.get_account_settle_profitloss?
````

```
api.get_account_settle_profitloss(
        product_type="0", #商品類型，預設為0全部
        summary="Y", #是否為匯總資料
        start_date="", #起始日期
        end_date="", #結束日期
        currency="", #幣別
        account={}, #交易帳戶
    )
```

### AccountSettleProfitLoss

````python
# 查看近180天已實現損益
from datetime import date, timedelta
st_date = (date.today() - timedelta(days=180)).strftime('%Y%m%d')
print(st_date)
settle_profitloss = api.get_account_settle_profitloss(summary='Y', start_date=st_date)
settle_profitloss
````

````python
df_profitloss = pd.DataFrame(settle_profitloss.data())
df_profitloss
````

````python
settle_profitloss = api.get_account_settle_profitloss(start_date='20211101', end_date='20211210')
df_settle_profitloss = pd.DataFrame(settle_profitloss.data())
df_settle_profitloss
````
